### PERGUNTA 4 do BB (1,50 pontos)

Para a variável `duration` considerando **apenas o perfil de clientes descrito no teste do Blackboard**, faça:

 > 1. Construção de um **histograma**.
 >
 > 1. Uso, obrigatoriamente, de **faixas iguais** com amplitudes de 60 segundos, iniciando em zero e terminando em 420 (incluindo esse valor 420 na última faixa).
 >
 > 1. Uso de **densidade**.

<br>


**Atenção:**  
 * Na Pergunta 4 do teste do Blackboard, assinale a interpretação correta a partir do histograma construído.


```python
# ESCREVA SEU CÓDIGO AQUI
```

```python
faixa1 = arange(0,480,60)
```

```python
df_no = df.duration.loc[ df.y == 'no']
df_no
```

```python
plt.figure(figsize = (15, 5))

plt.subplot(121)
plt.hist(df_no, bins = faixa1, edgecolor = 'orange')
plt.title('Contagem (eixo y) e Amplitudes iguais')
plt.ylabel('Frequência')

plt.subplot(122)
plt.hist(df_no, bins = faixa1, edgecolor = 'orange', density= True)
plt.title('Densidade (eixo y) e Amplitudes iguais')
plt.ylabel('Densidade')

plt.show()
```

```python
df_no.describe().to_frame()
```

```python
# Densidade = frequencia relativa/amplitude

df_NO = pd.cut(df_no, bins = faixa1, right=False)
```

```python
df_NO.value_counts()
```
