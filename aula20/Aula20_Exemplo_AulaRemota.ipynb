## Problema - Aulas remotas

Nas √©pocas de aulas presenciais, a distribui√ß√£o de probabilidades das notas em determinada prova historicamente se comportava como uma normal com m√©dia igual a 6 e desvio padr√£o igual a 0,8. 

Devido ao novo formato de alunas remotas, alguns especialistas desconfiam que a nota m√©dia nessa prova tenha aumentado. Ainda, segundo esses especialistas, n√£o existem raz√µes para acreditar em altera√ß√µes no valor do desvio padr√£o, na distribui√ß√£o de probabilidades da vari√°vel de interesse e nem que a m√©dia tenha diminu√≠do.

Como verificar se os especialistas est√£o certos?


```python
# Vari√°vel de interesse --> X: Nota de cada aluno nas aulas remotas
# Par√¢mtro de interesse --> E(X) = mu

# sigma = 0.8 (Desvio padr√£o)
# n = 36 alunos

# D√∫vida: 

# Nota m√©dia aula remota diminuiu nota m√©dia aula presencial --> mu <=6 : H0
# Nota m√©dia aula remota aumentou nota m√©dia aula presencial --> mu > 6 : H1 ou HA


# x_barra ~ N(E(X_barra) = mu ; Var(X_barra) = sigma**2/n )

# Sob H0: x_barra ~ N(6; sigma**2/n )
# Sob H0: x_barra ~ N(6; 0.8**2/n )
```

**Continua√ß√£o...**

Supondo ùêª_0 verdadeira, ou seja, ùúá = 6 e selecionando uma amostra aleat√≥ria de ùëõ = 36 provas.
Encontre ùëê tal que ùõº seja igual a 5%.

```python
# 1) vari√°vel de interesse  -->  X: nota de cada aluno
#    parametro de interesse -->  E(x) = mu
# 
# 2) D√∫vida nota m√©dia aula remota > nota media aula presencial ?
#    duas possibilidades:
#    nota media da aula remota n√£o aumentou nota m√©dia aula presencial ---> mu <= 6 hipotese nula ---------- H0
#    nota media da aula remota aumentou nota m√©dia da aula presencial  ---> mu >  6 hipotese alternatica --- H1
# 
# 3) Sob H0, Xbarra ~ N(6,0.8¬≤/n) enunciado diz que dp(X) = 0.8 e n = 36 alunos

# 4) Desenhar o gr√°fico sob H0

# 5) Conclus√£o - Seja c o ponto de corte, que defini o erro, temos que:
#                X_barra > c ---> rejeitamos H0
#           
```

```python
# sob a hipotese nula: H0
# H0: mu = 6 'Representa em termos do problema que remoto √© igual ao presencial'
# H1: mu > 6 <= pergunta
```

```python
n = 36
mu = 6
sigma = 0.8
alpha = 0.05 # n√≠vel de significancia ou P(erro I) = P(rejeitar H0| H0 √© verdadeiro)
sigma_Xbarra = sigma/sqrt(n)

ponto_c = stats.norm.ppf(1 - alpha, loc = mu, scale = sigma_Xbarra)
ponto_c
```

```python
print(f'Se a m√©dia amostral for superior ao ponto de corte = {ponto_c:.2f}, ent√£o rejeitaremos H0')
print('---> ou seja, a aula remota tem nota m√©dia superior ao das aulas presencias')
```
